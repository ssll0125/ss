# Pharmacokinetics of Perfluoroalkyl acids (PFAAs)

PFAAs are chemicals that are used for the catalysis of various chemical processes including polymerization and oligomerization. Domestically they are used by the automotive, aviation, electronic, packaging and textile industries. They are also actively used for biomaterial and household product development. They are present in biocides [Sznajder-Katarzy≈Ñska., 2019](https://www.hindawi.com/journals/jchem/2019/2717528/). Especially in the field of biomedical engineering, beginning from the preparation of gels to medical devices, perfluoroalkyl components and fluorocarbons are used [Krafft et al., 1998](https://pdf.sciencedirectassets.com/272115/1-s2.0-S0300908400X00569/1-s2.0-S0300908400800164/main.pdf?). 

PFAAs are a significant environmental concern due to their ability to accumulate in humans and animals. However, there is significant debate ongoing about the risk of outright bans on PFAAs, specifically for the biomedical community (see [this](https://nap.nationalacademies.org/catalog/26156/guidance-on-pfas-exposure-testing-and-clinical-follow-up) report by the National Academies, and [this](https://www.govinfo.gov/content/pkg/FR-2022-07-13/pdf/2022-14862.pdf) Request for Information from the White House).

In this ongoing effort to understand PFAAs and their impact on our bodies, one approach to study and predict the levels of PFAAs in humans is through the use of Pharmacokinetic models. 

Pharmacokinetics is a description of the concentration changes of a chemical or a drug in the body over time. Pharmacokinetics are most commonly used to describe the changes in concentrations in the blood, but there are very complex models out there (called Physiology Based Pharmacokinetic or PBPK models for those that are interested) that also account for concentration changes in various organs and tissues. 

In this project we will only consider concentrations in blood.

In this project, you will develop and use a simple pharmacokinetic model to describe the concentration of a PFAA in human blood. Through the various tasks, you will add increasing complexity and use your model to answer fundamental questions about the accumulation of PFAAs in blood.

This project is loosely based on the work by [Verner et al., 2016](https://pubs.acs.org/doi/10.1021/acs.est.5b04399) who evaluated the accumulation of these chemicals in mothers and newborn children.

There are a number of processes that can affect the pharmacokinetics of a chemical. Below is a non-exhaustive list that we will consider within this project:

1. **Exposure or dose**. This is the 'source' of the chemical. It can be a one-time exposure (like taking a pill) or it can be continuous (like from some environmental source in your house).

2. **Excretion**. If the chemical can be effectively removed from our bodies (e.g. through the kidneys). This represents a 'loss' of the chemical.

3. **Metabolism**. If the chemical can be converted into something else, this also represents a 'loss'. These are often enzymatic reactions where the chemical is converted to another substance that can then be excreted or used in a different way.

In the text cell below, solve the differential mass balance equation analytically, and use the analytical solution to answer the question.

Step 1: Starting Mass Balance Equation

$ \frac{dM}{dt} = k_e - k_{exc} \cdot M $    

Step 2: Separation of Variables

$ \frac{dM}{(k_e - k_{exc} \cdot M)} = dt$

Step 3: Integrate

$ -\frac{1}{k_{exc}}(\ln(k_e - k_{exc} \cdot M) - \ln(k_e)) = t$

Step 4: Plug in values

$ -\frac{1}{0.2}(\ln(90 - 0.2 \cdot (80 \cdot 5.5)) - \ln(90)) = t$

Final result:

$ t = 19.0333 $ seconds

In this task, you will **numerically** solve the differential mass balance equation that you derived in Task 1b. More specifically your code should do the following:

2. Define initial conditions for concentration and time
3. Define step size (h) and number of steps (n)

5. Set the first value of each array equal to the initial condition of the variable

#### Enter your code here

In the code cell below, write your code to achieve the goals that we outlined above.

```python
# Import libraries   
%matplotlib inline
import matplotlib.pyplot as plt

def dCdt_man(C): #Concentration function for man
  V_man = 5.5 #Blood volume of average man's body in liters
  k_e = 90 #Exposure rate in ng/day
  k_exc = 0.2 #rate of excretion in 1/day

  dCdt_man = (k_e/V_man) - k_exc * C  #Change in concentration over time for man

  return dCdt_man; 

def dCdt_woman(C): #Concentration function for woman
  V_woman = 4.5 #Blood volume of average woman's body in liters
  k_e = 90 #Exposure rate in ng/day
  k_exc = 0.2 #rate of excretion in 1/day
  dCdt_woman = (k_e/V_woman) - k_exc * C  #Change in concentration over time for woman

  return dCdt_woman

# Initial conditions
C0 = 0
t0 = 0
# Total length of time to solve over 
tf = 30
# Define the size of the timesteps that you want 
del_t = .1
# number of steps needed
num_steps = int((tf - t0)/del_t)

# Arrays for both concentrations and time
Cman_mesh = [0]*(num_steps + 1)
Cwoman_mesh = [0]*(num_steps + 1)
time_mesh = [0]*(num_steps + 1)

# Setting first index of array equal to initial conditions
Cman_mesh[0] = C0
Cwoman_mesh[0] = C0
time_mesh[0] = t0

# A foor loop that will calculate the concentration for both men and woman at the current time step and add it to the arrays
for j in range(num_steps):         # repeats until an index (j) reaches the value of num_steps

    Cman_mesh[j+1] = Cman_mesh[j] + dCdt_man(Cman_mesh[j])*del_t   
    Cwoman_mesh[j+1] = Cwoman_mesh[j] + dCdt_woman(Cwoman_mesh[j])*del_t    
    # update the time_mesh by adding del_t to the previous timestep
    time_mesh[j+1] = time_mesh[j] + del_t

# Plot both concentrations as a function of time
plt.plot(time_mesh,Cman_mesh,color='b') #function for man
plt.plot(time_mesh,Cwoman_mesh,color='r') #function for woman
plt.title('Concentration in Body Over Time') #plot title
plt.xlabel('Time (sec)') # x-axis (time in seconds)
plt.ylabel('Concentration (ng/L)') # y-axis (concentraion in ng/L)
plt.legend(['Man','Woman'])
plt.grid()

```

```python
# Import libraries   
%matplotlib inline
import matplotlib.pyplot as plt

def dCdt_man(C_p,C_e): #Concentration of PFAA function for man
  V_man = 5.5 #Blood volume of average man's body in liters
  k_e = 90 #Exposure rate in ng/day
  k_exc = 0.2 #rate of excretion in 1/day
  k_rxn = 0.02 #rate of reaction between PFAA and metabolite

  dCdt_man = (k_e/V_man) - (k_exc * C_p) - (k_rxn * C_p * C_e)  #Change in concentration over time for man

  return dCdt_man; 

def dCdt_woman(C_p,C_e): #Concentration of PFAA function for man
  V_woman = 4.5 #Blood volume of average man's body in liters
  k_e = 90 #Exposure rate in ng/day
  k_exc = 0.2 #rate of excretion in 1/day
  k_rxn = 0.02 #rate of reaction between PFAA and metabolite

  dCdt_woman = (k_e/V_woman) - (k_exc * C_p) - (k_rxn * C_p * C_e)  #Change in concentration over time for man

  return dCdt_woman; 

def dEdt_man(C_p,C_e):
  V_man = 5.5
  k_rxn = 0.02
  k_p = 10
  k_deg = 0.1

  dEdt_man = (k_p/V_man) - (k_deg * C_e) - (k_rxn * C_p * C_e)

  return dEdt_man

def dEdt_woman(C_p,C_e):
  V_woman = 4.5
  k_rxn = 0.02
  k_p = 10
  k_deg = 0.1

  dEdt_woman = (k_p/V_woman) - (k_deg * C_e) - (k_rxn * C_p * C_e)

  return dEdt_woman

# Initial conditions
C0_p = 0
C0_e_man = 100/5.5
C0_e_woman = 100/4.5
t0 = 0
# Total length of time to solve over 
tf = 30
# Define the size of the timesteps that you want 
del_t = .1
# number of steps needed
num_steps = int((tf - t0)/del_t)

# Arrays for both concentrations and time
Cman_mesh_p = [0]*(num_steps + 1)
Cwoman_mesh_p = [0]*(num_steps + 1)
Cman_mesh_e = [0]*(num_steps + 1)
Cwoman_mesh_e = [0]*(num_steps + 1)
time_mesh = [0]*(num_steps + 1)

# Setting first index of arrays equal to initial conditions
Cman_mesh_p[0] = C0_p
Cwoman_mesh_p[0] = C0_p
Cman_mesh_e[0] = C0_e_man
Cwoman_mesh_e[0] = C0_e_woman
time_mesh[0] = t0

# A foor loop that will calculate the concentration for both men and woman at the current time step and add it to the arrays
for j in range(num_steps):         # repeats until an index (j) reaches the value of num_steps

    Cman_mesh_p[j+1] = Cman_mesh_p[j] + dCdt_man(Cman_mesh_p[j],Cman_mesh_e[j])*del_t   
    Cwoman_mesh_p[j+1] = Cwoman_mesh_p[j] + dCdt_woman(Cwoman_mesh_p[j],Cwoman_mesh_e[j])*del_t 
    Cman_mesh_e[j+1] = Cman_mesh_e[j] + dEdt_man(Cman_mesh_p[j],Cman_mesh_e[j])*del_t    
    Cwoman_mesh_e[j+1] = Cwoman_mesh_e[j] + dEdt_woman(Cwoman_mesh_p[j],Cwoman_mesh_e[j])*del_t  
    # update the time_mesh by adding del_t to the previous timestep
    time_mesh[j+1] = time_mesh[j] + del_t

# Plot both concentrations as a function of time

plt.figure()
plt.plot(time_mesh,Cman_mesh_p,color='b') #function for man for PFAA
plt.plot(time_mesh,Cwoman_mesh_p,color='r') #function for woman for PFAA
plt.title('Concentration of PFAA in Body Over Time') #plot title
plt.xlabel('Time (sec)') # x-axis (time in seconds)
plt.ylabel('Concentration (ng/L)') # y-axis (concentraion in ng/L)
plt.legend(['Man','Woman'])
plt.grid()

plt.figure()
plt.plot(time_mesh,Cman_mesh_e,color='g') #function for man for E
plt.plot(time_mesh,Cwoman_mesh_e,color='r') #function for woman for E
plt.title('Concentration of E in Body Over Time') #plot title
plt.xlabel('Time (sec)') # x-axis (time in seconds)
plt.ylabel('Concentration (ng/L)') # y-axis (concentraion in ng/L)
plt.legend(['Man','Woman'])
plt.grid()

plt.show()


```
